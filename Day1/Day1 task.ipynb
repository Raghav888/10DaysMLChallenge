```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("time_series_2019-ncov-Confirmed.csv")
database=data.copy()
aa=[]
bb=[]
for i in range(4,65):
    m=database.iloc[:,i].value_counts()
    if(m[0]>341):
          aa.append(i)
data.drop(data.columns[aa],axis=1,inplace=True)
data=data.rename(columns={"Province/State":"State","Country/Region":"Country"})
```

```python
asia = pd.read_csv("as_countries.csv")
eu = pd.read_csv("eu_countries.csv")
af = pd.read_csv("af_countries.csv")
na = pd.read_csv("na_countries.csv")
sa = pd.read_csv("sa_countries.csv")
as_count = asia['name'].values.tolist()
eu_count = eu['name'].values.tolist()
af_count = af['name'].values.tolist()
na_count = na['name'].values.tolist()
sa_count = sa['name'].values.tolist()
```

```python
cases = data["3/18/20"]
countries = data["Country"]
```

```python
ind = 0
as_cases = 0
eu_cases = 0
af_cases = 0
na_cases = 0
sa_cases = 0
for i in countries:
    if i in as_count:
        as_cases+=cases[ind]
    if i in eu_count:
        eu_cases+=cases[ind]
    if i in af_count:
        af_cases+=cases[ind]
    if i in na_count:
        na_cases+=cases[ind]
    if i in sa_count:
        sa_cases+=cases[ind]
    ind=ind+1
x=[as_cases,eu_cases,af_cases,na_cases,sa_cases]
y=["Asia","Europe","Africa","NA","SA"]
plt.bar(y,x,color="RED")
```
